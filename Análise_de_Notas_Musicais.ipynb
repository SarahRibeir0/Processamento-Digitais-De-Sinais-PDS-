```python
from google.colab import drive
from IPython.display import Audio
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
caminho_notas = ["/content/Nota01.wav",
                 "/content/Nota02.wav",
                 "/content/Nota03.wav",
                 "/content/Nota04.wav"]
```

```python
for nota in caminho_notas:
  audio_data, sr = librosa.load(nota)
  display(Audio(audio_data, rate=sr, autoplay=True))
```

```python

audio_data_nota1, sample_rate_nota1 = librosa.load('/content/Nota01.wav')

audio_data_nota2, sample_rate_nota2 = librosa.load('/content/Nota02.wav')

audio_data_nota3, sample_rate_nota3 = librosa.load('/content/Nota03.wav')

audio_data_nota4, sample_rate_nota4 = librosa.load('/content/Nota04.wav')

```

```python
# Plotando Grfico da nota 1
plt.figure(figsize=(12, 4))
librosa.display.waveshow(audio_data_nota1, sr=sample_rate_nota1)
plt.title('Forma de Onda da Nota 1')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.show()

fft_nota1 = np.fft.fft(audio_data_nota1)
magnitude_spectrum_nota1 = np.abs(fft_nota1)
frequencia_nota1 = np.fft.fftfreq(len(audio_data_nota1), 1/sample_rate_nota1)

magnitude_spectrum_nota1 = magnitude_spectrum_nota1[:len(magnitude_spectrum_nota1)//2]
frequencia_nota1 = frequencia_nota1[:len(frequencia_nota1)//2]

fundamental_index = np.argmax(magnitude_spectrum_nota1)

fundamental_frequency = frequencia_nota1[fundamental_index]

print("Frequncia fundamental da Nota 1:", fundamental_frequency, "Hz")
print("Nota LA - A")

plt.plot(frequencia_nota1, magnitude_spectrum_nota1)
plt.title('Espectro de Frequncias da Nota 1')
plt.xlabel('Frequncia (Hz)')
plt.ylabel('Magnitude')
plt.show()
```
